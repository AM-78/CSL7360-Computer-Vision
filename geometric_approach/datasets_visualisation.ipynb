```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
folder = './datasets/simple_clocks/analog_clocks/images/'
# img = cv2.imread(folder + "0.jpg")
# img.hconcat(img)

rows = []
for row in range(4):
    one_row = []
    for col in range(4):
        one_row.append(cv2.imread(folder + f"{4*row + col}.jpg"))
    one_row = cv2.hconcat(one_row)
    rows.append(one_row)

img = cv2.vconcat(rows)
plt.axis("off")
plt.imshow(img)
plt.title("Synthetic Clocks Dataset")
plt.show()
```

```python
folder = './datasets/simple_clocks/simple_analog_clock_monochrome/clock_minute_hand/samples/'
files = os.listdir(folder)[:20]

rows = []
for row in range(4):
    one_row = []
    for col in range(4):
        one_row.append(cv2.imread(folder + files[4*row + col]))
    one_row = cv2.hconcat(one_row)
    rows.append(one_row)

img = cv2.vconcat(rows)

plt.axis("off")
plt.imshow(img)
plt.title("Monochrome Clock Dataset")
plt.show()
```

```python
folder = './datasets/augmented/TIME_Image_Dataset_Classification/train/'

subfolders = os.listdir(folder)[:50]
images = []

for i in subfolders:
    imgs_in_subfolder = os.listdir(folder + i)
    for j in imgs_in_subfolder:
        images.append(folder + i + "/" + j)

# print(images)

from random import shuffle

shuffle(images)

rows = []
for row in range(4):
    one_row = []
    for col in range(4):
        one_row.append(cv2.imread(images[4*row + col]))
    one_row = cv2.hconcat(one_row)
    # print(one_row)
    rows.append(one_row)

img = cv2.vconcat(rows)

# print(img)
plt.axis("off")
plt.imshow(img)
plt.title("TIME Dataset (Augmented using Rotation)")
plt.show()
```
